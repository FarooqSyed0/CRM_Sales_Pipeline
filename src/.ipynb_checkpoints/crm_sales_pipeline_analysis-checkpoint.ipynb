# CRM & Sales Pipeline Analysis

This notebook analyzes the **CRM-and-Sales-Pipelines.csv** dataset to answer the following impact questions:

1. **Pipeline Snapshot** – Current sales pipeline breakdown (number of opportunities at each stage).
2. **Sales Forecast** – Monthly forecast based on open opportunities, deal values, probabilities, and expected close dates.
3. **Opportunity Closure Time** – Average time from lead acquisition to close; comparison across countries.
4. **Actual Win Rates** – Overall win ratio and variations by country.
5. **Win Rates by Product and Region** – Win ratios across products and geographic markets.


## 1. Data Loading & Initial Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv('CRM-and-Sales-Pipelines.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

## 3. Pipeline Snapshot

```python
# ---- Pipeline Snapshot ----
# Use 'status' as the pipeline stage if 'stage' is missing
stage_col = np.where(df['stage'].notna(), df['stage'], df['status'])
df['pipeline_stage'] = stage_col

# Current opportunities are those not yet closed (no actual_close_date) OR status in specific stages
open_mask = df['actual_close_date'].isna()
pipeline_snapshot = df.loc[open_mask].groupby('pipeline_stage').size().sort_values(ascending=False)
print('Current Pipeline Snapshot (Open Opportunities):')
print(pipeline_snapshot)

# Plot
plt.figure(figsize=(10,4))
ax = sns.barplot(x=pipeline_snapshot.values, y=pipeline_snapshot.index, palette='viridis')
plt.title('Open Opportunities by Pipeline Stage')
plt.xlabel('Count')
plt.ylabel('Stage')
for i, v in enumerate(pipeline_snapshot.values):
    ax.text(v+1, i, str(v), color='black', va='center')
plt.tight_layout()
plt.show()
```

## 4. Sales Forecast

```python
# ---- Sales Forecast ----
# Consider open opportunities with an expected close date
open_ops = df[(df['actual_close_date'].isna()) & (df['expected_close_date'].notna())]

# Forecast value = deal_value * probability%
open_ops['forecast_value'] = open_ops['deal_value'] * (open_ops['probability'] / 100)

# Group by year-month of expected close
open_ops['close_month'] = open_ops['expected_close_date'].dt.to_period('M')
forecast = open_ops.groupby('close_month')['forecast_value'].sum().reset_index()
forecast['close_month'] = forecast['close_month'].astype(str)
print('Sales Forecast (by Expected Close Month):')
print(forecast.head())

# Plot
plt.figure(figsize=(12,4))
sns.lineplot(data=forecast, x='close_month', y='forecast_value', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales Forecast')
plt.ylabel('Forecasted Revenue')
plt.xlabel('Expected Close Month')
plt.tight_layout()
plt.show()
```

## 5. Opportunity Closure Time

```python
# ---- Opportunity Closure Time ----
closed_ops = df[df['actual_close_date'].notna()]
closed_ops['days_to_close'] = (closed_ops['actual_close_date'] - closed_ops['lead_acquisition_date']).dt.days

overall_avg = closed_ops['days_to_close'].mean()
print(f'Overall average days to close: {overall_avg:.1f}')

country_closure = closed_ops.groupby('country')['days_to_close'].mean().sort_values()
print('Average Days to Close by Country:')
print(country_closure)

# Plot
plt.figure(figsize=(10,4))
ax = sns.barplot(x=country_closure.values, y=country_closure.index, palette='coolwarm')
plt.title('Avg Days to Close by Country')
plt.xlabel('Days')
plt.ylabel('Country')
for i, v in enumerate(country_closure.values):
    ax.text(v+1, i, f'{v:.1f}', color='black', va='center')
plt.tight_layout()
plt.show()
```

## 6. Actual Win Rates

```python
# ---- Actual Win Rates ----
# Define a win as stage == 'Won' OR status == 'Customer'
win_mask = (df['stage'] == 'Won') | (df['status'] == 'Customer')
loss_mask = (df['stage'] == 'Lost') | (df['status'] == 'Churned Customer') | (df['status'] == 'Disqualified')

win_count = df[win_mask].shape[0]
loss_count = df[loss_mask].shape[0]
win_rate = win_count / (win_count + loss_count)
print(f'Overall Win Rate: {win_rate:.2%}')

# Win rate by country
country_summary = df.assign(win=win_mask.astype(int), loss=loss_mask.astype(int))
country_win = country_summary.groupby('country')['win'].sum()
country_total = country_summary.groupby('country').apply(lambda x: (x['win'] + x['loss']).sum())
country_win_rate = (country_win / country_total).dropna().sort_values(ascending=False)
print('Win Rate by Country:')
print(country_win_rate)

# Plot
plt.figure(figsize=(10,4))
ax = sns.barplot(x=country_win_rate.values, y=country_win_rate.index, palette='Greens')
plt.title('Win Rate by Country')
plt.xlabel('Win Rate')
plt.xlim(0,1)
plt.ylabel('Country')
for i, v in enumerate(country_win_rate.values):
    ax.text(v+0.02, i, f'{v:.1%}', color='black', va='center')
plt.tight_layout()
plt.show()
```

## 7. Win Rates by Product and Region

```python
# ---- Win Rates by Product and Region ----
summary = country_summary.copy()
summary['product'] = df['product']

product_region = (summary.groupby(['product','country'])[['win','loss']].sum())
product_region['win_rate'] = product_region['win'] / (product_region['win'] + product_region['loss'])
print(product_region['win_rate'].unstack().fillna(0).round(2))
```
